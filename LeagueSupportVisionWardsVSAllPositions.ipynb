4. What is Crosstabs and what are they used for?
```
Crosstabs are helpful for creating new 
DataFrames from an existing Dataframe; e.g. 
if we had a DataFrame describing aspects of 
different participants of a poll which we 
will call Voting Data, let's say that each 
row is a unique person, and the columns are 
things like Population (with distinct 
values being rv for registered voter, lv 
for likely voter, and gen for local 
population), Age (stating their age to the 
nearest whole number), Political Part, etc. 
Instead of looking at the rows one by one 
to get a general gist of the population, we 
can use crosstab to create a table with 
Political Party as rows and Population as 
the column to concisely get an idea of  the 
count of each type of population for each 
political party. Crosstabs are useful for 
arranging data in a more readible and 
summarized format than just a raw table.
```
5. What is the difference between a T-Test and a P-Test?
```
A T-Test gives evidence of whether or not 
two sample means are significantly 
different from each other; since the 
measure used to perform the T-test is in 
units of standard error (or the standard 
deviation of the sample), the greater the 
magnitude of T, the greater the evidence 
against your null hypothesis that the two 
samples are similar and/or dependent; the 
lower the t-value, the more likely that 
there is not a significant difference 
between the two sample means.
```
```
A P-Test helps measure the probability of a 
t-value; the more extreme the t-value, the 
lower the p-value will be since such 
extreme t-value are highly unlikely to 
happen. Also, this helps determine if the 
outcome of the hypothesis test is random or 
legitimate.
```
